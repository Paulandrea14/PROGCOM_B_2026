Ejercicio 5

Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

```python
estudiantes = {"Ana": 4.5, "Luis": 3.5, "Maria": 5.0, "Carlos": 3.0}
def promedio_general(d):
    suma=0
    cantidad=0
    for nota in d.values():
        suma=suma+nota
        cantidad=cantidad+1
    return suma/cantidad
def promedio_estudiante(d, nombre):
    if nombre in d:
        return d[nombre]
    else:
        return "No existe"
print("Promedio general:", promedio_general(estudiantes))
print("Promedio de Ana:", promedio_estudiante(estudiantes, "Ana"))
```

Ejercicio 8

Dada una lista de elementos recolectados en una expedición, cuenta cuántas veces aparece cada elemento.

Lista sugerida: ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]

```python
elementos=["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo={}
for elemento in elementos:
    if elemento in conteo:
        conteo[elemento] = conteo[elemento] + 1
    else:
        conteo[elemento] = 1
print(conteo)
```

Ejercicio 10

Simula un pequeño inventario.
Debes permitir:

agregar un objeto,
actualizar la cantidad de un objeto,
eliminar un objeto,
mostrar el inventario final.

```python
inventario = {}
obj = input("Agregar objeto: ")
inventario[obj] = int(input("Cantidad: "))
obj = input("Actualizar objeto: ")
while obj not in inventario:
    obj = input("Ese objeto no existe, intenta otra vez: ")
inventario[obj] = int(input("Nueva cantidad: "))
obj = input("Eliminar objeto: ")
while obj not in inventario:
    obj = input("Ese objeto no existe, intenta otra vez: ")
del inventario[obj]
print("\nInventario final:")
for k, v in inventario.items():
    print(k, ":", v)
```
