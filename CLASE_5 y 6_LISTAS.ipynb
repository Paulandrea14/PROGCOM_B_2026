#WHILE
1. Cuando no se sabe cuántas veces se repite una acción
2. Es gobernado por una condición de parada
3. Es necesaria una condición inicial, la cuál **DEBE** cambiar adentro del ciclo.

Sintaxis
```
while (condición):
  código a repetir
  modificación de condición
```

**OJO**:
La modificación de la condición puede darse de diferentes formas:
1. Contador
2. Un acumulador
3. Cambio de bandera (trabajando con booleanos)
4. Con un input

#FUNCIONES
Es un pedazo de código "muerto" que no se ejecuta hasta que el código principal lo ejecute.

Ventanas
1. Organiza el código
2. Habilita un nuevo paradigma de programación: Programación modular
3. Habilita el copiado dinámico (reutilización del código)

Sintaxis
```
def nombre_funcion(p1,p2,p3,...,pn):
  aquí va el código
  return (dato)
```
Donde:
* p1, p2, ... pn son parámetros
* return retorna un dato

Existen 4 tipos diferentes de funciones
1. Sin parámetro, sin retorno
2. Sin parámetro, con retorno
3. Con parámetro, sin retorno
4. Con parámetro, con retorno
