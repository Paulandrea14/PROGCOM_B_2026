17. Escribe un programa que calcule la suma de números naturales hasta n.

```python
#FOR
suma = 0
num=int(input("Ingrese el número: "))
for i in range(1, num + 1):
    suma+=i
print("La suma de los números naturales es:",suma)
```

```python
#WHILE
suma=0
num=int(input("Ingrese un número: "))
i = 1
while i <= num:
    suma+=i
    i=i+1
print("La suma de los números naturales e:", suma)
```

18. Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

```python
#FOR
num=int(input("Ingrese un número: "))
if num < 0:
  print("No existen factoriales negativos")
else:
  factorial=1
  for i in range(1, num + 1):
      factorial=factorial*i
  print("El factorial es:", factorial)
```

```python
#WHILE
num = int(input("Ingrese un número: "))
if num<0:
  print("No existen factoriales negativos")
else:
  factorial=1
  i=1
  while i <= num:
      factorial=factorial*i
      i=i+1
  print("El factorial es:", factorial)
```

19. Crea un programa que muestre los primeros n números de la serie Fibonacci.

```python
#FOR
a=0
b=1
num=int(input("¿Cuántos números de la serie Fibonacci quieres ver? "))
for i in range(num):
  print(a)
  c=a+b
  a=b
  b=c
```

20. Escribe un programa que adivine entre un conjunto de diez colores definidos por el usuario, con un máximo de 10 intentos.

```python
#FOR
colores=[]
for i in range(10):
  color=input("Escribe un color: ")
  colores.append(color)
secreto = colores[0]
for i in range(10):
  intento=input("Adivina el color: ")
  if intento==secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
print("El color era:", secreto)
```

```python
#WHILE
colores=[]
i=0
while i < 10:
  color = input("Escribe un color: ")
  colores.append(color)
  i=i+1
secreto = colores[0]
intentos = 0
while intentos < 10:
  intento = input("Adivina el color: ")
  if intento == secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
  intentos = intentos + 1
print("El color era:", secreto)
```

21. Desarrolla un programa que muestre los números pares del 1 al 100.

```python
#FOR
for i in range(1, 101):
  if i % 2 == 0:
    print(i)
```

```python
#WHILE
i=1
while i <= 100:
  if i % 2 == 0:
    print(i)
  i=i+1
```

22. Crea un programa que calcule la suma de los dígitos de un número.

```python
#FOR
suma=0
num=input("Ingresa un número: ")
for digito in num:
    suma+=int(digito)
print("La suma de los dígitos es:", suma)
```

```python
#WHILE
suma=0
num=int(input("Ingresa un número: "))
while num>0:
  digito=num % 10
  suma+=digito
  num=num//10
print("La suma de los dígitos es:", suma)
```

23. Escribe un programa que determine si un número es primo.

```python
#FOR
contador=0
num=int(input("Ingresa un número: "))
for i in range(1, num + 1):
  if num % i == 0:
      contador = contador + 1
if contador == 2:
    print("El número es primo")
else:
    print("El número no es primo")
```

24. Desarrolla un programa que muestre el patrón de un triángulo con asteriscos.

```python
#FOR
n=int(input("Ingrese la altura del triángulo: "))
for i in range(1, n + 1):
    print("*" * i)
```
