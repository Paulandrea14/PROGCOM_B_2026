```python
def mover(cadera, veces):
  for i in range(veces):
    for j in range(cadera):
      print("Muevo cadera derecha")
    for k in range(cadera):
      print("Muevo cadera izquierda")
```

```python
mover(2,4)
mover(1,4)
```
