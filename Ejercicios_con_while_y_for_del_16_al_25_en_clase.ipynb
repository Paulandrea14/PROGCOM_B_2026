16. Crea un programa que muestre la tabla de multiplicar de un número hasta el 10.

```python
#FOR
print("La tabla del 2 es:")
for i in range(1,11):
  print(f"2*{i}: {i*2}")
```

```python
#WHILE
c=1
print("Tabla del 2")
while c<=10:
  print(f"2x{c}={2*c}")
  c+=1
```
