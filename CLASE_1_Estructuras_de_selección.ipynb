##IF/ELSE
Evalúa una o más condiciones, y dependiendo de su valor de verdad, realizará una acción u otra (solo dos acciones)

Sintaxis


```
if (condición):
 Si la condición es verdadera, se ejecuta este pedazo de código
else:
 Si la condición es falsa, se ejecuta este pedazo de código
```



````python
lluvia=input("Está lloviendo?")
if lluvia=="sí" or lluvia=="si" or lluvia=="sí. Muchísimo":
  print("Use sombrilla. Guarde la ropa")
else:
  print("Salga tranquilo")
````

## if ternario
Sintaxis
```
variable= (código si condición es verdadera)
if (condición) else (código si condición es falsa)
```



````python
aguacero="Use sombrilla. Guarde la ropa" if lluvia=="si" else "Salga tranquilo"
print(aguacero)
````
