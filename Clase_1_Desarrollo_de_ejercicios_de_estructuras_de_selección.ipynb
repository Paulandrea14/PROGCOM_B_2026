1. Escribe un programa que calcule el área de un triángulo pidiendo base y altura al usuario.


```python
base=float(input("Escriba la base de del triángulo: "))
altura=float(input("Escriba la altura del triángulo: "))
área=(base*altura)/2
print(área)
```

4. Escribe un programa que calcule el perímetro y área de un círculo dado su radio.

5. Crea un programa que resuelva una ecuación cuadrática (ax² + bx + c = 0) pidiendo los valores de a, b y c.

```python
a=float(input("Ingrese el valor de a: "))
b=float(input("Ingrese el valor de b: "))
c=float(input("Ingrese el valor de c: "))
ecuacióncuadrática1=(-b+((b**2)-(4*a*c)**0.5))/2*a
ecuacióncuadrática2=(-b-((b**2)-(4*a*c)**0.5))/2*a
print("El valor de la primera solución de la ecuación cuadrática es:",ecuacióncuadrática1)
print("El valor de la segunda solución de la ecuación cuadrática es:",ecuacióncuadrática2)
```

7. Crea un programa que determine si un año es bisiesto (divisible por 4 y no por 100, o divisible por 400).

```python
año=float(input("Ingrese el año: "))
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
  print("El año es bisiesto")
else:
  print("El año no es bisiesto")
```
