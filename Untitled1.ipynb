##IF/ELSE
Se usa cuando una condición solo puede dar dos valores de verdad: Verdadero y falso
Sintaxis en python
```
if (condición):
  Aquí se ejecuta código si la condición es verdadera
else:
 Aquí se ejcuta código si la ejecución es falsa
```
Sintaxis en java
```
if (condición){
  Aquí se ejecuta código si la condición es verdadera
}else{
  Aquí se ejecuta código si la condición es falsa
}
```




````python
edad=int(input("Cuál es tu edad"))
if edad>= 18:
  print("Usted es mayor de edad")
else:
  print("Usted es menor de edad")
````
