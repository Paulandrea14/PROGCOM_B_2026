#WHILE
1. Cuando no se sabe cuántas veces se repite una acción
2. Es gobernado por una condición de parada
3. Es necesaria una condición inicial, la cuál **DEBE** cambiar adentro del ciclo.

Sintaxis
```
while (condición):
  código a repetir
  modificación de condición
```

**OJO**:
La modificación de la condición puede darse de diferentes formas:
1. Contador
2. Un acumulador
3. Cambio de bandera (trabajando con booleanos)
4. Con un input

#LISTAS
1. Tienen index (numeración)
2. Son finitas
3. Son mutables

Sintaxis
```
# Forma canónica
variable=[]
# Extensión
variable=[e,e1,e2,e3,...,en]
# Comprensión: fórmula que genera todos los elementos de la lista.
```



````python
figura=["angelitos","estrella fugaz","dinosaurios","diablitos"]
cantidad=[4,2,5,3]
````

##Operaciones con listas
1. Recorrer listas
2. Verificar elementos dentro de la lista
3. Borrar elementos en la lista
4. Agregar elementos en la lista
5. Mutabilidad de las listas

###1. Recorrer listas

````python
#len:Calcula el tamaño de un dato
for elemento in range(len(figura)):
  print(figura[elemento])
````

````python
for elemento in range(len(figura)):
  print(figura[elemento],":", cantidad[elemento])
````

````python
for elemento in figura:
  print(elemento)
````

###2. Verificar elementos dentro de la lista


````python
user=input("Qué figura de galleta quiere comprar?")
for elemento in range(len(figura)):
  if user==figura[elemento]: print("Si hay galleta")
  else: print("No hay galleta")
````

````python
if user in figura: print("Si hay galleta")
else: print("No hay galleta")
````

````python
print("Si hay galleta" if user in figura else "No hay galleta")
````
